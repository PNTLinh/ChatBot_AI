```python
import os, sys, shutil
import numpy as np
from sentence_transformers import SentenceTransformer
from langchain_community.vectorstores import FAISS

base = "/content/drive/MyDrive/Colab Notebooks/ChatBot_AI/"
sys.path.insert(0, base + "src")
from data_loader import load_procedures

docs, metas = load_procedures(base + "data/TTHC.json")
print(f"Loaded {len(docs)} documents")

store_dir = base + "data/faiss_store"
if os.path.isdir(store_dir):
    print(f"Removing existing store at {store_dir}")
    shutil.rmtree(store_dir)

# Sử dụng sentence-transformers trực tiếp
print("Loading the model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Generating embeddings...")
embeddings = model.encode(docs)

# Tạo FAISS vectorstore
print("Building the vectorstore...")
vectorstore = FAISS.from_texts(
    texts=docs,
    embedding=lambda x: model.encode([x])[0],  # Cung cấp hàm embedding trực tiếp
    metadatas=metas
)

vectorstore.save_local(store_dir)
print(f"Vectorstore built and saved to {store_dir}")
```
